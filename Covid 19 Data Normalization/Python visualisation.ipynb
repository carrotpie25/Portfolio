```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\ASUS\Documents\Michelle\Avery\Sem 2\DTS106\question5b.csv',delimiter='\t')

# Combine 'year', 'month', and 'day' into one column
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

# Exclude December 2019 for a neat graph
df = df[(df['date'].dt.year != 2019) | (df['date'].dt.month != 12)]

plt.figure(figsize=(9, 5))

# Plot the cumulative deaths for each country
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.plot(country_data['date'], country_data['cumulative_deaths'], label=country)

plt.xlabel('Date')
plt.ylabel('Cumulative Deaths')
plt.title('Cumulative Deaths Over Time for Top 10 Countries')

# Limit the x axis to show only months
unique_months = df['date'].dt.to_period('M').unique()
plt.xticks(unique_months.to_timestamp(), [month.strftime('%b %Y') for month in unique_months], rotation=45)

plt.show()
```
